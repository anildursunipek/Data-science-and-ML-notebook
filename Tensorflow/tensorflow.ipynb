# Data split Test-Train

```python
from sklearn.model_selection import train_test_split
# y -> Label
y = df['Price'].values
# x -> Features
x = df[['Year','Km']].values

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_train2 = pd.DataFrame(x_train)
x_train2.describe()
```

```python
from tensorflow.keras.models import Sequential #Model class
from tensorflow.keras.layers import Dense # Model layer
```

```python
model = Sequential()
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1)) # output layer

model.compile(optimizer="rmsprop",loss="mse") #Cost Function algorithm
```

```python
model.fit(x_train,y_train,epochs=250) #Epochs
```

```python
loss = model.history.history["loss"]
model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLose = model.evaluate(x_train,y_train,verbose=0)
```

```python
testLose = model.evaluate(x_test,y_test,verbose=0)
```

```python
testLose
```

```python
trainLose
```

```python
testPredict = model.predict(x_test)
testPredict
```

```python
predictTestDf = pd.DataFrame(y_test,columns=["Real Values"])
```

```python
testPredict = pd.Series(testPredict.reshape(330,))
testPredict
```

```python
predictDf
```
