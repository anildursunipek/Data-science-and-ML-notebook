```python
y = df["Type"].values #Label
```

```python
x = df.drop("Type",axis=1).values #Features
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
lossDf = pd.DataFrame(model.history.history)
lossDf.head()
```

```python
lossDf.plot()#Overfitting
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modeLoss = pd.DataFrame(model.history.history)
modeLoss.plot()
```

```python
model = Sequential()            #DROPOUT

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
predict = (model.predict(x_test) > 0.5).astype("int")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```
